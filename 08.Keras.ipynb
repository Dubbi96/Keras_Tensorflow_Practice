```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from tensorflow.keras.layers import Dense
from tensorflow.keras import Sequential
from tensorflow.keras.optimizers import Adam
```

```python
df = pd.read_csv('./data/data/cars.csv')
df.columns = ['speed','dist']
df
```

```python
#units = label(output)의 종류 / input_dim = feature(input)의 종류
IO = Dense(units=1, input_dim = 1)
model = Sequential([IO])
model.compile(loss = 'mean_squared_error', optimizer = Adam(0.1))
h = model.fit(df['speed'], df['dist'],epochs =50)
```

```python
h.history['loss']
```

```python
plt.plot(h.history['loss'])
plt.show()
```
