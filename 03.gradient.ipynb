```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
```

```python
def cost(x,y,w):
    c = 0
    for i in np.arange(len(x)):
        hx = w*x[i]
        c = c + (hx - y[i])**2
        return c/len(x)
```

```python
def gradient(x,y,w):
    c = 0 ##initialize c
    for i in np.arange(len(x)):
        hx = w*x[i]
        c = c + (hx - y[i])*x[i] ##편미분값
        return c/len(x)
```

```python
def show_gradient():
    x = [1,2,3]
    y = [1,2,3]
    w = 10
    for i in range(200):
        c = cost(x,y,w)
        g = gradient(x,y,w)
        w = w-0.1*g
        print(i,c)
    print("w:",w)
```

```python
show_gradient()
```
