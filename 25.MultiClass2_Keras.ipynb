```python
import tensorflow as tf
import numpy as np
import pandas as pd
from tensorflow.keras.layers import Dense
from tensorflow.keras import Sequential
from tensorflow.keras.optimizers import Adam
from sklearn import datasets
#from tensorflow.keras.backend import one_hot
from tensorflow.keras.utils import to_categorical # one-hot keras encording
import matplotlib.pyplot as plt
```

```python
iris = datasets.load_iris()
iris
```

```python
iris['target']
```

```python
ydata = to_categorical(iris['target'],3)
```

```python
xdata = iris['data'] #sl 5.1, sw 3.5, pl 1.4, pw 0.2 종분류 및 정확도
xdata
```

```python
IO = Dense(units=3, input_dim=4, activation='softmax') #sigmoid를 활성함수로 쓰겠다. 
model = Sequential( [IO] ) #순차모델
model.compile( loss='categorical_crossentropy', optimizer=Adam(0.1),metrics=['accuracy'])  #loss는 로그 함수를 사용한다. 
h = model.fit( xdata, ydata, epochs=1000)
```

```python
model.predict(np.array([[5.1,3.5,1.4,0.2]]))
```

```python
model.predict_classes(np.array([[5.1,3.5,1.4,0.2]]))
```

```python
IO.get_weights()
```

```python
h.history['acc'][-1]
```

```python
pp = model.predict(xdata)
(pp.argmax(axis =1 ) == iris['target']).mean()
```
