## 패션이미지에 따른 상품을 예측하시요(텐서, 케라스)

```python
(train_images,train_labels),(test_images,test_labels) = load_data()
```

```python
sess=tf.Session()
```

```python
class_names=['T-shirt/top','Trouser','Pullover','Dress','Coat','Sandal','Shirt','Sneaker','Bag','Ankle boot']

#(train_images,train_labels),(test_images,test_labels) 나누어져 있음 
```

```python
train1=one_hot(train_labels,10)
sess.run(train1)
```

```python
train2=one_hot(test_labels,10)
sess.run(train2)
```

```python
x_train = train_images
x_test = test_images
y_train = train1
y_test = train2
```

```python
x_train = x_train.reshape(-1, 28*28)
```

```python
x_test = x_test.reshape(-1,28*28)
```

```python
x = tf.placeholder( tf.float32, [None,784]) #28X28묶여있음
y = y_train
```

```python
w1 = tf.get_variable( 'w1',[784,100] ,
        initializer=tf.contrib.layers.xavier_initializer() )
b1 = tf.get_variable( 'b1',[100] ,
        initializer=tf.contrib.layers.xavier_initializer() )
w2 = tf.get_variable( 'w2',[100,10] ,
        initializer=tf.contrib.layers.xavier_initializer() )
b2 = tf.get_variable( 'b2',[10] ,
        initializer=tf.contrib.layers.xavier_initializer() )
```

```python
z1 = tf.matmul( x, w1) + b1
lay1 = tf.nn.relu( z1 )
z2 = tf.matmul( lay1, w2 ) + b2
hx = tf.nn.softmax( z2 )
```

```python
cost_i = tf.nn.softmax_cross_entropy_with_logits_v2(logits=z2,
                                            labels=y)
cost = tf.reduce_mean(cost_i)
```

```python
optimizer = tf.train.AdamOptimizer(0.1)#learning rate
train = optimizer.minimize( cost )
sess = tf.Session()
init =tf.global_variables_initializer()
sess.run( init)
```

```python
for i in range(1001):
    sess.run( train, {x:x_train})
    print( i, sess.run(cost, {x:x_train}))
```
