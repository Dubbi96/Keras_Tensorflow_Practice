```python
x = tf.placeholder(tf.float32,[None,4])
y = y_train
w = tf.Variable(tf.random_uniform([4,3])) 
b = tf.Variable(tf.random_uniform([3]))
```

```python
z = tf.matmul(x,w)+b
hx = tf.nn.softmax(z)
cost_i = tf.nn.softmax_cross_entropy_with_logits_v2(logits = z ,labels = y)
cost = tf.reduce_mean(cost_i)
```

```python
optimizer = tf.train.GradientDescentOptimizer(0.1)
train = optimizer.minimize(cost)
sess = tf.Session()
init = tf.global_variables_initializer()
sess.run(init)
```

```python
for i in range(3000):
    sess.run(train,{x:x_train})
    c = sess.run(cost,{x:x_train})
    print(i,c)
```

```python
t = sess.run(hx,{x:x_test})

result = t.argmax(axis =1)

val = y_test.argmax(axis=1)
```
