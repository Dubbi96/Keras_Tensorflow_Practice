```python
from keras.models import Sequential
from keras.layers.core import Dense
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import StratifiedKFold
from sklearn import model_selection

import numpy as np
import pandas as pd
import tensorflow as tf

# seed 값 설정
seed = 0
numpy.random.seed(seed)
tf.random.set_seed(3)
```

## 1. 데이터 읽어오기

```python
df = pd.read_csv('C:/Users/user/ILIFO/DeepLearning/deeplearning_class-master/dataset/sonar.csv', header=None)
df
```

## 2. 데이터 전처리

```python
dataset = df.values
X = dataset[:,0:60]
Y_obj = dataset[:,-1]
```

```python
# 문자열 변환
e = LabelEncoder()
e.fit(Y_obj)
Y = e.transform(Y_obj)
```

```python
# 학습셋과 테스트셋을 나눔
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.3, random_state=seed)
X_train1 = np.asarray(X_train)
Y_train1 = np.asarray(Y_train)
X_test1 = np.asarray(X_test)
Y_test1 = np.asarray(Y_test)
```

## 3. 모델 설계

```python
model = Sequential()
model.add(Dense(24, input_dim=60, activation='relu'))
model.add(Dense(10, activation='relu'))
model.add(Dense(1,  activation='sigmoid'))
```

## 4. 모델 컴파일

```python
model.compile(loss = 'binary_crossentropy', optimizer = "Adam" ,metrics = ['accuracy'])  #optimizer = "adam" 으로 설정하면 xavier 초기화
```

## 5. 학습

```python
h= model.fit(X_train1,Y_train1,epochs=130,batch_size=5)
```
